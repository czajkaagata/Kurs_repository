```python
""" Zadanie 7 – Zwracanie wielu wartości z funkcji
Napisz funkcję 
analiza_liczb(lista) która zwraca krotkę z: sumą, średnią, minimum i
maksimum z listy. Przetestuj na liście 
Wymagania:
[10, 20, 30, 40, 50] .
Funkcja zwraca krotkę z 4 wartościami
Użyj tuple unpacking przy wywołaniu
Wypisz wyniki"""
def analiza_liczb(lista):
    suma = sum(lista)
    srednia = suma / len(lista)
    minimum = min(lista)
    maksimum = max(lista)
    return suma, srednia, minimum, maksimum

wyniki = analiza_liczb([10, 20, 30, 40, 50])
suma, srednia, minimum, maksimum = wyniki
print(f"Suma: {suma}")
print(f"Średnia: {srednia}")
print(f"Minimum: {minimum}")
print(f"Maksimum: {maksimum}")
```

```python
"""✏ Zadanie 9 – Inwersja słownika
Masz słownik mapujący nazwy produktów na kategorie: {'laptop': 'elektronika',
'klawiatura': 'elektronika', 'krzesło': 'meble', 'biurko': 'meble'} . Stwórz
odwrotny słownik gdzie klucze to kategorie, a wartości to listy produktów.
Wymagania:
Oczekiwany wynik:
(średnie)
Iteruj po oryginalnym słowniku
Zgrupuj produkty według kategorii
Wynik: {'elektronika': ['laptop', 'klawiatura'], 'meble': [...]}
Słownik kategoria: lista_produktów"""
słownik = {'laptop': 'elektronika', 'klawiatura': 'elektronika', 'krzesło': 'meble', 'biurko': 'meble'}
odwrotny = {}   
for produkt, kategoria in słownik.items():
    if kategoria not in odwrotny 
        odwrotny[kategoria] = []
        odwrotny[kategoria].append(produkt) 
print(odwrotny)
```

```python
"""10. ✏ Zadanie 10 – Walidacja danych ze zbiorem dozwolonych wartości
Masz listę kategorii od użytkowników: ['sport', 'muzyka', 'SPORT', 'film', 'xyz',
'nauka', 'gaming'] . Dozwolone kategorie to {'sport', 'muzyka', 'film', 'nauka',
'gry'} . Znajdź nieprawidłowe kategorie (uwzględnij wielkość liter - konwertuj na małe).
Wymagania:
(średnie)
Konwertuj wszystkie kategorie na małe litery
Użyj operacji zbiorowych by znaleźć nieprawidłowe
Wypisz prawidłowe i nieprawidłowe osobno"""
kategorie_uzytkownikow = {'sport', 'muzyka', 'SPORT', 'film', 'xyz', 'nauka', 'gaming'}
dozwolone_kategorie = {'sport', 'muzyka', 'film', 'nauka', 'gry'}
kategorie_uzytkownikow = set(kategoria.lower() for kategoria in kategorie_uzytkownikow)
nieprawidlowe = kategorie_uzytkownikow - dozwolone_kategorie
prawidlowe = kategorie_uzytkownikow & dozwolone_kategorie
print("Prawidłowe kategorie:", prawidlowe)
print("Nieprawidłowe kategorie:", nieprawidlowe)
```

```python
"""12. ✏ Zadanie 12 – Analiza transakcji - słowniki zagnieżdżone
Masz listę transakcji (krotek): [('2024-01-15', 'Elektronika', 1200), ('2024-01-
16', 'Meble', 800), ('2024-01-15', 'Elektronika', 350), ...] . Stwórz słownik
gdzie kluczem jest data, a wartością słownik z sumą sprzedaży dla każdej kategorii tego
dnia.
Dataset:
Wymagania:
Oczekiwany wynik:
(średnie)
transakcje = [
('2024-01-15', 'Elektronika', 1200),
('2024-01-15', 'Meble', 450),
('2024-01-16', 'Elektronika', 800),
('2024-01-16', 'Elektronika', 350),
('2024-01-15', 'Odzież', 200),
('2024-01-16', 'Meble', 600)
]
Zagnieżdżone słowniki: {data: {kategoria: suma}}
Iteruj po transakcjach i akumuluj sumy
Słownik typu: {'2024-01-15': {'Elektronika': 1200, 'Meble': 450, ...}, ...} """

transakcje = [
('2024-01-15', 'Elektronika', 1200),    
('2024-01-15', 'Meble', 450),    
('2024-01-16', 'Elektronika', 800),    
('2024-01-16', 'Elektronika', 350),    
('2024-01-15', 'Odzież', 200),    
('2024-01-16', 'Meble', 600)
]
wynik = {}
for data, kategoria, kwota in transakcje:
    if data not in wynik:
        wynik[data] = {}
    if kategoria not in wynik[data]:
        wynik[data][kategoria] = 0
    wynik[data][kategoria] += kwota

print(wynik)
```
