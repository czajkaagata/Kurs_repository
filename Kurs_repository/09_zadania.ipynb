 Zadanie 1 – Tworzenie klasy Produkt
Stwórz klasę 
Produkt z atrybutami: 
nazwa , 
Dodaj metody:
cena , 
ilosc_w_magazynie .
pokaz_info() – wyświetla informacje o produkcie
sprzedaj(ilosc) – zmniejsza 
ilosc_w_magazynie o podaną wartość
dostawa(ilosc) – zwiększa 
ilosc_w_magazynie
Stwórz 3 obiekty i przetestuj metody.

```python
class Produkt:
    def __init__(self, nazwa, cena, ilosc_w_magazynie):
        self.nazwa = nazwa
        self.cena = cena
        self.ilosc_w_magazynie = ilosc_w_magazynie

    def pokaz_info(self):
        print(f"Produkt: {self.nazwa}, Cena: {self.cena}, Ilość w magazynie: {self.ilosc_w_magazynie}")

    def sprzedaj(self, ilosc):
        if ilosc <= self.ilosc_w_magazynie:
            self.ilosc_w_magazynie -= ilosc
            print(f"Sprzedano {ilosc} sztuk {self.nazwa}.")
        else:
            print(f"Niewystarczająca ilość {self.nazwa} w magazynie.")

    def dostawa(self, ilosc):
        self.ilosc_w_magazynie += ilosc
        print(f"Dostarczono {ilosc} sztuk {self.nazwa}.")
```

Stwórz obiekt i wykonaj przykładowe obliczenia.

```python
class Kalkulator:
    def podziel(self, a, b):
        if b != 0:
            return a / b
        else:
            return "Nie można dzielić przez zero."

    def dodaj(self, a, b):
        return a + b   
kalkulator = Kalkulator()
print(kalkulator.dodaj(5, 3)) 
print(kalkulator.podziel(10, 2))  
print(kalkulator.podziel(10, 0))   
```

 Zadanie 3 – Klasa BankAccount
Stwórz klasę 
BankAccount z atrybutami: 
odejmij(a, b) , 
pomnoz(a, b) ,
wlasciciel , 
saldo .
Metody:
wplac(kwota) – zwiększa saldo
wyplac(kwota) – zmniejsza saldo (tylko jeśli wystarczy środków)
pokaz_saldo() – wyświetla saldo
Stwórz kilka kont i przetestuj operacje.
(proste)

```python
class BankAccount:
    def __init__(self, wlasciciel: str, saldo: int=0 ):
        self.wlasciciel = wlasciciel
        self.saldo = saldo

    def wplac(self, kwota: int):
        self.saldo += kwota
        print(f"Wpłacono {kwota}. Nowe saldo: {self.saldo}")

    def wyplac(self, kwota: int)-> bool:
        if kwota <= self.saldo:
            self.saldo -= kwota
            print(f"Wypłacono {kwota}. Nowe saldo: {self.saldo}")
            return True
        else:
            print("Niewystarczające środki na koncie.")
            return False

    def pokaz_saldo(self):
        print(f"Saldo konta {self.wlasciciel}: {self.saldo}")
konto1 = BankAccount("Jan Kowalski", 1000)
konto1.pokaz_saldo()
konto1.wplac(500)      
konto1.wyplac(200)
konto1.pokaz_saldo()
konto2 = BankAccount("Anna Nowak", 2000)
konto2.pokaz_saldo()
```

Zadanie 9 – Pipeline z wieloma krokami
Stwórz 4 komponenty:
FiltrowanieDodatnich – usuwa liczby ujemne
Zaokraglanie – zaokrągla do całkowitych
Skalowanie – mnoży przez stały współczynnik
Sumowanie – zwraca sumę listy
Stwórz klasę 
(średnie)
Pipeline i przetestuj na danych: 
[1.7, -3.2, 5.8, -1.1, 8.3, 2.4]

```python
class FiltrowanieDodatnich:
    def przetworz(self, dane):
        return [x for x in dane if x >= 0]
class Zaokraglanie:
    def przetworz(self, dane):
        return [round(x) for x in dane]
class Skalowanie:
    def __init__(self, wspolczynnik):
        self.wspolczynnik = wspolczynnik
    def przetworz(self, dane):
        return [x * self.wspolczynnik for x in dane]
class Sumowanie:
    def przetworz(self, dane):
        return sum(dane)  


class Pipeline:
    def __init__(self, komponenty):
        self.komponenty = komponenty
    def przetworz(self, dane):
        for komponent in self.komponenty:
            dane = komponent.przetworz(dane)
        return dane 
dane = [1.7, -3.2, 5.8, -1.1, 8.3, 2.4]
pipeline = Pipeline([
    FiltrowanieDodatnich(),
    Zaokraglanie(),
    Skalowanie(10),
    Sumowanie()
])
wynik = pipeline.przetworz(dane)
print(f"Wynik {wynik}")
```

 Zadanie 10 – Klasa Zamowienie z kompozycją
Stwórz klasę 
Produkt (nazwa, cena) oraz klasę 
Zamowienie , która zawiera listę
produktów.
Metody 
Zamowienie :
dodaj_produkt(produkt, ilosc)
laczna_wartosc() – suma wartości zamówienia
pokaz_zamowienie() – wyświetla wszystkie produkty i łączną wartość
Stwórz zamówienie z 5 produktami

```python
class Produkt:
    def __init__(self, nazwa, cena):
        self.nazwa = nazwa
        self.cena = cena    
class Zamowienie:
    def __init__(self):
        self.produkty = []

    def dodaj_produkt(self, produkt, ilosc):
        self.produkty.append((produkt, ilosc))

    def laczna_wartosc(self):
        return sum(produkt.cena * ilosc for produkt, ilosc in self.produkty)

    def pokaz_zamowienie(self):
        print("Zamówienie:")
        for produkt, ilosc in self.produkty:
            print(f"{produkt.nazwa:<20} | {ilosc:<5} | {produkt.cena:>10.2f} zł")
        print(f"Łączna wartość: {self.laczna_wartosc()}")

produkt1 = Produkt("Laptop", 3000)
produkt2 = Produkt("Smartphone", 1500) 
produkt3 = Produkt("Słuchawki", 200)
produkt4 = Produkt("Monitor", 800)
produkt5 = Produkt("Klawiatura", 100)

zamowienie = Zamowienie()
zamowienie.dodaj_produkt(produkt1, 1)
zamowienie.dodaj_produkt(produkt2, 2)
zamowienie.dodaj_produkt(produkt3, 3)
zamowienie.dodaj_produkt(produkt4, 1)
zamowienie.dodaj_produkt(produkt5, 4)

zamowienie.pokaz_zamowienie()
```

Zadanie 11 – System ocen uczniów
Stwórz klasy:
Uczen (imie, oceny=[])
Klasa (nazwa, uczniowie=[])
Metody 
Klasa :
dodaj_ucznia(uczen)
srednia_klasy() – średnia ocen wszystkich uczniów
najlepszy_uczen() – zwraca ucznia z najwyższą średnią
Stwórz klasę z 5 uczniami i oblicz statystyki.
(średnie)

```python
class Uczen:
    def __init__(self, imie, oceny=None):
        self.imie = imie
        self.oceny = oceny if oceny is not None else []

    def srednia(self):
        if self.oceny:
            return sum(self.oceny) / len(self.oceny)
        return 0

class Klasa:
    def __init__(self, nazwa):
        self.nazwa = nazwa
        self.uczniowie = []

    def dodaj_ucznia(self, uczen):
        self.uczniowie.append(uczen)

    def srednia_klasy(self):
        if self.uczniowie:
            return sum(uczen.srednia() for uczen in self.uczniowie) / len(self.uczniowie)
        return 0

    def najlepszy_uczen(self):
        if self.uczniowie:
            return max(self.uczniowie, key=lambda uczen: uczen.srednia())
        return None

uczen1 = Uczen("Jan", [1, 2, 3])
uczen2 = Uczen("Anna", [5, 5, 1])

klasa = Klasa("3A")
klasa.dodaj_ucznia(uczen1)
klasa.dodaj_ucznia(uczen2)
 
print(f"Średnia  {klasa.nazwa}: {klasa.srednia_klasy():.2f}")

najlepszy = klasa.najlepszy_uczen()
if najlepszy:
    print(f"Najlepszy uczeń: {najlepszy.imie} ze średnią {najlepszy.srednia():.2f}")
        
         
```

 Zadanie 12 – Dziedziczenie – FiguraGeometryczna
Stwórz klasę bazową 
Figura z metodą 
pole() .
Stwórz klasy potomne:
Prostokat(szerokosc, wysokosc)
Kolo(promien)
Każda klasa implementuje 
pole() .
Stwórz listę figur i oblicz łączne pole.
(średnie)

```python
""" Zadanie 12 – Dziedziczenie – FiguraGeometryczna
Stwórz klasę bazową 
Figura z metodą 
pole() .
Stwórz klasy potomne:
Prostokat(szerokosc, wysokosc)
Kolo(promien)
Każda klasa implementuje 
pole() .
Stwórz listę figur i oblicz łączne pole.
(średnie)"""
import math
class Figura:
    def pole(self):
        raise NotImplementedError("Metoda pole() musi być zaimplementowana w klasie potomnej.")
class Prostokat(Figura):
    def __init__(self, szerokosc, wysokosc):
        self.szerokosc = szerokosc
        self.wysokosc = wysokosc

    def pole(self):
        return self.szerokosc * self.wysokosc

class Kolo(Figura):
    def __init__(self, promien):
        self.promien = promien

    def pole(self):
        return math.pi * self.promien ** 2

figury = [
    Prostokat(5, 3),
    Kolo(2),
    Prostokat(4, 6)
]

lacznepole = sum(figura.pole() for figura in figury)
print(f"Łączne pole {lacznepole}")
```
