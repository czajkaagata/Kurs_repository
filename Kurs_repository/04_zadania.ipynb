```python
""" Zadanie 1 – Kalkulator BMI
Napisz program, który obliczy wskaźnik BMI (Body Mass Index).
Dane:
Waga: 70 kg
Wzrost: 1.75 m
Wymagania:
Stwórz zmienne 
waga i 
wzrost
Oblicz BMI według wzoru: BMI = waga / (wzrost²)
Wyświetl wynik z 2 miejscami po przecinku
Dodaj interpretację: BMI < 18.5 (niedowaga), 18.5-24.9 (norma), ≥25 (nadwaga)
Oczekiwany rezultat:
BMI: 22.86
Interpretacja: Norma"""
waga = 70
wzrost = 1.75
bmi = waga / (wzrost ** 2)
print(f'BMI: {bmi:.2f}')
if bmi < 18.5:
    interpretacja = "Niedowaga"
elif 18.5 <= bmi < 25:
    interpretacja = "Norma"     
else:
    interpretacja = "Nadwaga"   
print(f'Interpretacja: {interpretacja}')
```

```python
""" Zadanie 2 – Konwersja temperatur
Napisz program konwertujący temperatury Celsius na Fahrenheit i Kelvin.
Dane:
Temperatura: 23.5°C
Wymagania:
Fahrenheit = Celsius × 9/5 + 32
Kelvin = Celsius + 273.15
Wyświetl wszystkie 3 wartości sformatowane z 1 miejscem po przecinku
Oczekiwany rezultat:
23.5°C = 74.3°F = 296.6 K"""
celsius = 23.5
fahrenheit = celsius * 9/5 + 32
kelvin = celsius + 273.15
print(f'{celsius}°C = {fahrenheit:.1f}°F = {kelvin:.1f} K') 
```

```python
"""Zadanie 3 – Podział zbioru danych
Oblicz rozmiary zbiorów treningowego, walidacyjnego i testowego.
Dane:
Całkowita liczba próbek: 1000
Proporcje: 70% trening, 15% walidacja, 15% test
Wymagania:
Użyj mnożenia i dzielenia całkowitego
Upewnij się, że suma = 1000 (obsłuż resztę z dzielenia)
Wyświetl rozmiary wszystkich zbiorów"""
calk_probek = 1000
procent_trening = 70
procent_walidacja = 15
procent_test = 15
rozmiar_trening = calk_probek * procent_trening // 100
rozmiar_walidacja = calk_probek * procent_walidacja // 100
rozmiar_test = calk_probek * procent_test // 100
reszta = calk_probek - (rozmiar_trening + rozmiar_walidacja + rozmiar_test)
rozmiar_trening += reszta  
print(f'Zbiór treningowy: {rozmiar_trening} próbek')
print(f'Zbiór walidacyjny: {rozmiar_walidacja} próbek')
print(f'Zbiór testowy: {rozmiar_test} próbek')  
```

```python
""" Zadanie 4 – Formatowanie nazw kolumn
Ustandaryzuj nazwę kolumny z datasetu.
Dane:
Nazwa raw: 
" PATIENT_age_YEARS "
Wymagania:
Usuń białe znaki
Zamień na małe litery
Zastąp _ przez 
Oczekiwany rezultat:
" "
Użyj title case
Patient Age Years"""
raw_name = "  PATIENT_age_YEARS  "
clean_name = raw_name.strip().lower().replace("_", " ").title()

print(f'Oryginał: "{raw_name}"')
print(f'Wynik: "{clean_name}"')
```

```python
"""Zadanie 6 – Dokładność w procentach
Sformatuj metryki modelu"""
data = {
    "Accuracy": 0.9234,
    "Precision": 0.8891,
    "Recall": 0.9456
    }

print(f"{'Metryka':<15} | {'Wartość':>10}")
print("-" * 28)

# Pętla - wyniki z wyrównaniem i formatowaniem
for name, value in data.items():
    # <15 wyrównanie do lewej
    # >10 wyrównanie do prawej
    # :.2% formatowanie procentowe z 2 miejscami po przecinku
    print(f"{name:<15} | {value:>10.2%}")
```

```python
"""✏
 Zadanie 9 – Normalizacja Min-Max
Znormalizuj wartość do zakresu [0, 1]"""
# Dane
min = 50
max = 100
wartosci = [75, 50, 100, 62.5]

def normalizuj(value, min_col, max_col):
    return (value - min_col) / (max_col - min_col)

# Nagłówek
print(f"{'Oryginał':^15} | {'Znormalizowana':^15}")
print("-" * 35)

for a in wartosci:
    norm = normalizuj(a, min, max)
    
    # Warunek
    if 0 <= norm <= 1:
        # TYLKO dla wartości spełniających warunek
        print(f"{a:^15} | {norm:^15.2f}")
    
```
