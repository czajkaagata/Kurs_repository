 Zadanie 9 – Standaryzacja datasetu
Wczytaj cały dataset Iris (4 cechy). Zaimplementuj standaryzację (mean=0, std=1) dla
każdej cechy używając operacji NumPy. Porównaj wynik z 
Dataset:
Iris (sklearn.datasets.load_iris)
StandardScaler z sklearn.
Wymagania:
Oblicz mean i std per kolumna (
Oczekiwany rezultat:
axis=0 )
Zastosuj wzór: (X - mean) / std
Sprawdź mean i std po standaryzacji
Porównaj z 
sklearn.preprocessing.StandardScaler
Oblicz max różnicę między implementacjami
Macierz X_standardized (150, 4)
Weryfikacja: mean≈0, std≈1
Max różnica vs sklearn

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

# wczytanie danych
iris = load_iris()
X = iris.data  # macierz cech (150, 4)

# mean i std dla kolumny
mean = np.mean(X, axis=0)
std = np.std(X, axis=0)
#print mean, std
print("Mean", np.round(mean, 6))
print("Std", np.round(std, 6))

# zastosowanie wzoru: (X - mean) / std
X_stand = (X - mean) / std


# sprawdzenie mean i std po standaryzacji

print("Mean", np.round(np.mean(X_stand, axis=0), 6))
print("Std", np.std(X_stand, axis=0))


# porównanie z sklearn.preprocessing.StandardScaler
scaler = StandardScaler()
X_sklearn = scaler.fit_transform(X)
# obliczenie max różnicy między implementacjami
max_roznica = np.max(np.abs(X_stand - X_sklearn))
print("Max różnica między implementacjami", max_roznica)
```

""" Zadanie 10 – Macierz odległości dla k-NN
Wybierz 10 losowych obserwacji z Iris. Oblicz pełną macierz odległości euklidesowych
(10×10) między wszystkimi parami. Wizualizuj jako heatmapę.
Dataset:
Iris (sklearn.datasets.load_iris)
Wymagania:
Losowo wybierz 10 obserwacji (użyj 
Użyj 
np.random.choice na indeksach)
euclidean_distances z sklearn lub zagnieżdżone pętle
Wizualizuj macierz jako heatmapę (seaborn.heatmap lub plt.imshow)
Diagonala powinna być 0 (odległość od samego siebie)
Oczekiwany rezultat:
Macierz 10×10
Heatmapa z annotacjami
(średnie"""


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.metrics import pairwise_distances
# wczytanie danych
iris = load_iris()
X = iris.data  # macierz cech (150, 4)

# losowy wybór 10 obserwacji
np.random.seed(42)  #zawsze te same wyniki
obserwacje = np.random.choice(X.shape[0], size=10, replace=False)
print("indeksy\n", obserwacje)

X_random = X[obserwacje]
print("cechy\n", X_random)
# macierz odległości euklidesowych
distance_matrix = pairwise_distances(X_random, metric='euclidean')

# wizualizacja
plt.figure(figsize=(8, 6))
sns.heatmap(distance_matrix, annot=True, fmt=".2f", cmap="viridis")
plt.title("Macierz odległości euklidesowych (10×10)")
plt.xlabel("Obserwacja")
plt.ylabel("Obserwacja")
plt.show()

```
