 Zadanie 2 – Konwerter jednostek
Napisz trzy funkcje: 
celsius_to_fahrenheit(c) , 
fahrenheit_to_celsius(f) ,
celsius_to_kelvin(c) .
Dataset:
Temperatury w Celsius: 
Wymagania:
[0, 10, 20, 30, 37, 100]
Każda funkcja zwraca skonwertowaną wartość
Zastosuj funkcje do konwersji wszystkich temperatur
Oczekiwany rezultat:
Tabela z trzema kolumnami: °C, °F, K
(proste)



```python
temp = [0, 10, 20, 30, 37, 100]

def celsius_to_fahrenheit(c):
    f = (c * 9/5) + 32
    return f
def fahrenheit_to_celsius(f):
    c = (f - 32) * 5/9
    return c
def celsius_to_kelvin(c):
    k = c + 273.15
    return k
print(f"{'°C':>15} | {'°F':>10} | {'K':>10}")
print("-" * 50)
for c in temp:
    f_val = celsius_to_fahrenheit(c)
    k_val = celsius_to_kelvin(c)
    print(f"{c:>15} | {f_val:>10.1f} | {k_val:>10.1f}")
```

Napisz funkcję 
powitanie(imie, jezyk="polski") , która wyświetla powitanie w
wybranym języku (polski, angielski, niemiecki).
Dataset:
Brak (funkcja operuje na stringach)
Wymagania:
Domyślny język to polski
Obsługa 3 języków
Wywołaj funkcję 3 razy z różnymi parametrami
Oczekiwany rezultat:
powitanie("Anna") → "Cześć, Anna!"
powitanie("Anna", "angielski") → "Hello, Anna!"

```python
def powitanie(imie, jezyk="polski"):
    if jezyk == "polski":
        return f"Cześć, {imie}!"
    elif jezyk == "angielski":
        return f"Hello, {imie}!"
    elif jezyk == "niemiecki":
        return f"Halo, {imie}!"
    else:
        return f"Witaj, {imie}!"
    
print(powitanie("Anna"))
print(powitanie("Anna", "angielski"))
```

Zadanie 8 – Map i filter
Używając 
map() i 
filter() z lambda, przetwórz listę liczb: najpierw zostaw tylko
dodatnie, potem podwój każdą.
Dataset:
liczby = [-5, 3, -2, 8, 0, -1, 7, 4, -9, 6]
Wymagania:
filter() z lambda: 
x > 0
map() z lambda: 
x * 2
Konwertuj wynik na listę
Oczekiwany rezultat:
Lista: 
[6, 16, 14, 8, 12] (tylko dodatnie i podwojone)

```python
liczby = [-5, 3, -2, 8, 0, -1, 7, 4, -9, 6]
dodatnie = filter(lambda x: x > 0, liczby) 
podwojone = map(lambda x: x * 2, dodatnie)
wynik = list(podwojone)
print(wynik)
```

Zadanie 9 – Funkcja normalizacji min-max
Napisz funkcję 
normalize_minmax(dane) , która normalizuje listę liczb do zakresu [0, 1]
używając metody min-max.
Dataset:
wynagrodzenia = [3500, 4200, 5800, 7200, 9500, 4800, 6100]
Wymagania:
Wzór: 
(x - min) / (max - min)
Funkcja zwraca nową listę znormalizowanych wartości
Zaokrąglij do 3 miejsc po przecinku
Oczekiwany rezultat:
Lista wartości w zakresie [0.0, 1.0

```python
#1 sztywne wynagrodzenia
wynagrodzenia = [3500, 4200, 5800, 7200, 9500, 4800, 6100]
normalize_minmax = lambda x: round((x - min(wynagrodzenia)) / (max(wynagrodzenia) - min(wynagrodzenia)),3)
normalize = list(map(normalize_minmax, wynagrodzenia))

print(normalize)

#2 niezależn dane 
normalize_minmax_uni = lambda x, dane: round((x - min(dane)) / (max(dane) - min(dane)), 2)
wynagrodzenia2 = [3500, 4200, 5800, 7200, 9500, 4800, 6100]
normalize_uni = list(map(lambda x: normalize_minmax_uni(x, wynagrodzenia2), wynagrodzenia2))

print(normalize_uni)
```

 Zadanie 10 – Dictionary comprehension - agregacja
Masz listę transakcji: 
[{"produkt": "X", "wartosc": 100}, ...] . Używając dict
comprehension i grupowania, oblicz łączną wartość dla każdego produktu.
Dataset:
transakcje = [
{"produkt": "laptop", "wartosc": 2999},
{"produkt": "mysz", "wartosc": 89},
{"produkt": "laptop", "wartosc": 2999},
{"produkt": "klawiatura", "wartosc": 249},
{"produkt": "mysz", "wartosc": 89},
{"produkt": "mysz", "wartosc": 89}
]
Wymagania:
Pogrupuj po produkcie
Zsumuj wartości dla każdego produktu
Wynik jako słownik: 
Oczekiwany rezultat:
{"laptop": 5998, ...}
Słownik z sumą wartości per produkt

```python
#1 petle
transakcje = [
{"produkt": "laptop", "wartosc": 2999},
{"produkt": "mysz", "wartosc": 89},
{"produkt": "laptop", "wartosc": 2999},
{"produkt": "klawiatura", "wartosc": 249},
{"produkt": "mysz", "wartosc": 89},
{"produkt": "mysz", "wartosc": 89}
]

def grupuj_i_sumuj(transakcje):
    wynik = {}
    for transakcja in transakcje:
        produkt = transakcja["produkt"]
        wartosc = transakcja["wartosc"]
        if produkt in wynik:
            wynik[produkt] += wartosc
        wynik[produkt] = wartosc
    return wynik
suma_per_produkt = grupuj_i_sumuj(transakcje)
print(suma_per_produkt)

#2 dict compre
transakcje = [
{"produkt": "laptop", "wartosc": 2999},
{"produkt": "mysz", "wartosc": 89},
{"produkt": "laptop", "wartosc": 2999},
{"produkt": "klawiatura", "wartosc": 249},
{"produkt": "mysz", "wartosc": 89},
{"produkt": "mysz", "wartosc": 89}
]
produkty = {t["produkt"] for t in transakcje}
print(produkty)
podsumowanie = {produkt: sum(t["wartosc"] for t in transakcje if t["produkt"] == produkt) for produkt in produkty}
print(podsumowanie)

#2 
```

Zadanie 11 – Feature engineering z comprehension
Masz dane klientów: wiek i dochód. Dodaj nowe cechy: 
średni/starszy) i 
kat_dochodu (niski/średni/wysoki).
Dataset:
klienci = [
{"id": 1, "wiek": 25, "dochod": 4500},
{"id": 2, "wiek": 45, "dochod": 9200},
{"id": 3, "wiek": 32, "dochod": 6800},
{"id": 4, "wiek": 28, "dochod": 5200},
{"id": 5, "wiek": 55, "dochod": 12000}
]
kat_wiekowa (młody/
Wymagania:
Kategoria wiekowa: <30 = młody, 30-50 = średni, >50 = starszy
Kategoria dochodu: <5000 = niski, 5000-10000 = średni, >10000 = wysoki
Użyj list comprehension do tworzenia nowej listy z dodatkowymi cechami
Oczekiwany rezultat:
Lista klientów z 2 nowymi cechami

```python
klienci = [
{"id": 1, "wiek": 25, "dochod": 4500},
{"id": 2, "wiek": 45, "dochod": 9200},
{"id": 3, "wiek": 32, "dochod": 6800},
{"id": 4, "wiek": 28, "dochod": 5200},
{"id": 5, "wiek": 55, "dochod": 12000}
]
def grupa_dochodowa (dochod):
    if dochod <5000:
        return "niski"
    elif dochod <10000:
        return "średni"
    else:
        return "wysoki"

def grupa_wiekowa (wiek):
    if wiek <30:
        return "młody"
    if wiek <=50:
        return "średni"
    else:
        return "starszy"

klienci = [{**klient, "kat_wiekowa": grupa_wiekowa(klient["wiek"]), "kat_dochodu": grupa_dochodowa(klient["dochod"])} for klient in klienci]
for klient in klienci:
    print(klient)
```
