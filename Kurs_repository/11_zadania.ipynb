Zadanie 1 – Tworzenie DataFrame z danych
Stwórz DataFrame zawierający informacje o 10 produktach w sklepie: nazwa, cena, ilość na
stanie, kategoria.
Wymagania:
Użyj słownika do utworzenia DataFrame
Wyświetl pierwsze 5 wierszy
Oblicz całkowitą wartość towaru (cena × ilość) i dodaj jako nową kolumnę
Oczekiwany rezultat:
DataFrame z 10 produktami i kolumną 'WartośćCałkowita

```python
import pandas as pd
# Tworzenie słownika z danymi produktów
produkty = {
    'Nazwa': ['Produkt A', 'Produkt B', 'Produkt C', 'Produkt D', 'Produkt E',
              'Produkt F', 'Produkt G', 'Produkt H', 'Produkt I', 'Produkt J'],
    'Cena': [10.99, 5.49, 20.00, 15.75, 8.99, 12.50, 7.25, 9.99, 14.30, 6.80],
    'IlośćNaStanie': [100, 200, 150, 80, 120, 90, 60, 110, 130, 70],
    'Kategoria': ['Elektronika', 'Dom', 'Elektronika', 'Ogród', 'Dom',
                  'Elektronika', 'Ogród', 'Dom', 'Elektronika', 'Ogród']
}
df = pd.DataFrame(produkty) 
# Wyświetlanie pierwszych 5 wierszy
print(df.head())
# Obliczanie całkowitej wartości towaru i dodawanie jako nową kolumnę
df['WartośćCałkowita'] = df['Cena'] * df['IlośćNaStanie']
# Wyświetlanie DataFrame z nową kolumną
print(df)
```

```python
"""Zadanie 5 – Wykrywanie NaN
Sprawdź które kolumny w Titanicu mają brakujące dane.
Wymagania:
Policz NaN w każdej kolumnie
✏ Zadania średnie (9-12)
(proste)
Oblicz procent NaN
Wyświetl tylko kolumny z >10% braków"""
# Sprawdzanie NaN w każdej kolumnie
nan_counts = titanic_df.isna().sum()
# Obliczanie procentu NaN w każdej kolumnie
nan_percentages = (nan_counts / len(titanic_df)) * 100
# Wyświetlanie tylko kolumn z >10% braków
columns_with_high_nans = nan_percentages[nan_percentages > 10]
print("Kolumny z większą niż 10% brakami danych:")
print(columns_with_high_nans)
```

Wypełnij brakujące wartości w kolumnie 'Age' średnią.
Wymagania:
(proste)
Oblicz średnią wieku (bez NaN)
Wypełnij NaN tą średnią

```python
# Obliczanie średniej wieku (bez NaN)
sredni_wiek_bez_nan = titanic_df['Age'].mean() 
# Wypełnianie NaN średnią
titanic_df['Age'].fillna(sredni_wiek_bez_nan, inplace=True)
# Sprawdzanie czy nie ma już NaN w 'Age'
nan_in_age = titanic_df['Age'].isna().sum()
print(f"Liczba NaN w kolumnie 'Age' po wypełnieniu: {nan_in_age}")
```

Wyczyść dataset Titanic stosując pełny pipeline:

```python
"""Zadanie 10 – Kompleksowe czyszczenie danych
Wyczyść dataset Titanic stosując pełny pipeline:
Wymagania:
(średnie)
Usuń kolumny z >60% NaN
Wypełnij Age medianą
Wypełnij Embarked dominantą
Usuń wiersze z pozostałymi NaN
Przed i po: wypisz kształt DataFrame
Sprawdź czy nie ma NaN
Porównaj rozkład Age przed i po
11. ✏ Zadanie 11 – Feature engineering
Stwórz nowe zmienne z Titanic:
Wymagania:
(średnie)
FamilySize = SibSp + Parch + 1
IsAlone = 1 jeśli FamilySize == 1, inaczej 0
AgeGroup = kategorie (0-12, 13-18, 19-35, 36-60, 60+)
Dodaj wszystkie 3 zmienne do DataFrame
Wyświetl value_counts() dla każdej
Oblicz wskaźnik przeżycia dla IsAlone vs nie alone"""
# Kompleksowe czyszczenie danych
print(f"Kształt DataFrame przed czyszczeniem: {titanic_df.shape}")
# Usuwanie kolumn z >60% NaN    
titanic_df.dropna(thresh=len(titanic_df) * 0.4, axis=1, inplace=True)
# Wypełnianie Age medianą
srednia_wiek = titanic_df['Age'].median()
titanic_df['Age'].fillna(srednia_wiek, inplace=True)
# Wypełnianie Embarked dominantą
dominanta_embarked = titanic_df['Embarked'].mode()[0]
titanic_df['Embarked'].fillna(dominanta_embarked, inplace=True)
# Usuwanie wierszy z pozostałymi NaN
titanic_df.dropna(inplace=True)
print(f"Kształt DataFrame po czyszczeniu: {titanic_df.shape}")
# Sprawdzanie czy nie ma NaN
nan_counts_after = titanic_df.isna().sum()
print("Liczba NaN w każdej kolumnie po czyszczeniu:")
print(nan_counts_after)
# Porównywanie rozkładu Age przed i po czyszczeniu
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
titanic_df['Age'].hist(bins=20, color='blue', alpha=0.7)
plt.title('Rozkład wieku po czyszczeniu')
plt.subplot(1, 2, 2)
titanic_df['Age'].hist(bins=20, color='orange', alpha=0.7)
plt.title('Rozkład wieku po czyszczeniu')
plt.tight_layout()
plt.show()
# Feature engineering
titanic_df['FamilySize'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1
titanic_df['IsAlone'] = (titanic_df['FamilySize'] == 1).astype(int)
titanic_df['AgeGroup'] = pd.cut(titanic_df['Age'], bins=[0, 12, 18, 35, 60, float('inf')],
                                  labels=['0-12', '13-18', '19-35', '36-60', '60+'])
# Wyświetlanie value_counts() dla każdej nowej zmiennej
print("Value counts dla FamilySize:")
print(titanic_df['FamilySize'].value_counts())
print("\nValue counts dla IsAlone:")
print(titanic_df['IsAlone'].value_counts())
print("\nValue counts dla AgeGroup:")
print(titanic_df['AgeGroup'].value_counts())
# Obliczanie wskaźnika przeżycia dla IsAlone vs nie alone
survival_rate_isalone = titanic_df.groupby('IsAlone')['Survived'].mean()
print("\nWskaźnik przeżycia dla IsAlone vs nie alone:")
print(survival_rate_isalone)
```
