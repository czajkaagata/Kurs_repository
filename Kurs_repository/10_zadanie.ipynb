Zadanie 9 – Normalizacja danych
Dataset: Oceny z 3 przedmiotów dla 20 studentów
Wymagania:
Oczekiwany rezultat:
(średnie)
scores = np.random.randint(40, 100, size=(20, 3))
Zastosuj normalizację Min-Max per przedmiot (kolumna) → [0, 1]
Sprawdź, że min=0 i max=1 dla każdego przedmiotu
Wizualizuj rozkład przed i po normalizacji (histogramy)
Tablica znormalizowana
2 histogramy (przed/po)

```python

scores = np.random.randint(40, 100, size=(20, 3))
print(scores)
min_vals = np.min(scores, axis=0)
max_vals = np.max(scores, axis=0)

normalized_scores = (scores - min_vals) / (max_vals - min_vals)
print(normalized_scores)

#sprawdzenie, że wartości są między 0 a 1 per przedmiot
print(np.min(normalized_scores, axis=0)) 
print(np.max(normalized_scores, axis=0))  

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(scores.flatten(), bins=10, color='blue', alpha=0.7)
plt.title('Przed normalizacją')
plt.subplot(1, 2, 2)
plt.hist(normalized_scores.flatten(), bins=10, color='green', alpha=0.7)
plt.title('Po normalizacji')    
plt.show()
```

exams = np.array([

Dodaj bonus do każdego egzaminu: [5, 10, 5, 10]
Następnie pomnóż wyniki każdego studenta przez mnożnik: [1.0, 1.1, 0.95, 1.15, 1.0]

```python
import numpy as np
exams = np.array([
    [78, 85, 92, 88],   #student 1
    [65, 72, 68, 70],
    [90, 88, 95, 92],
    [55, 60, 58, 62],
    [82, 79, 85, 80]
])
bonus = np.array([5, 10, 5, 10])
multipliers = np.array([1.0, 1.1, 0.95, 1.15, 1.0])
print("Oceny przed dodaniem bonusu i mnożeniem:")
print(exams)
# dodawanie bonus do każdego egzaminu i mnożenie wyników każdego studenta przez mnożnik
exams_final = (exams + bonus) * multipliers.reshape(5, 1)
print("Oceny po dodaniu bonusu i mnożeniu:")
print(exams_final)
```

Zadanie 11 – Analiza datasetu sprzedaży
Dataset: Sprzedaż 6 produktów w 30 dni
Oblicz:
Wymagania:
(średnie)
sales = np.random.randint(50, 200, size=(6, 30))
Trend sprzedaży per produkt (różnica między ostatnimi 7 dniami a pierwszymi 7 dniami)
Najlepszy tydzień (dni 0-6, 7-13, 14-20, 21-27) dla całej sprzedaży
Wykres: sprzedaż per produkt w czasie (linie)
Użyj slicingu do wycinania tygodni
Wizualizacja matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Generowanie datasetu
sales = np.random.randint(50, 200, size=(6, 30))
print("Dataset sprzedaży:")
print(sales)

# trend sprzedaży per produkt   
trend = sales[:, -7:] - sales[:, :7]
print("Trend sprzedaży:")
print(trend)  


# slicujemy wszystkie wiersze co skok 7 i sumujemy 
week_sales = np.array([sales[:, i:i+7].sum() for i in range(0, 28, 7)])
print("Sprzedaż w poszczególnych tygodniach:")
print (week_sales)

# najlepszy tydzień dla całej spredazy
best_week = np.argmax(week_sales) 
print("Najlepszy tydzień:")
print(f"Tydzień {best_week} z sumą sprzedaży: {week_sales[best_week]}")


# Wykres sprzedaży per produkt w czasie
days = np.arange(1, 31) 
plt.figure(figsize=(10, 6))
for i in range(sales.shape[0]):
    plt.plot(days, sales[i], label=f'Produkt {i+1}')
plt.xlabel('Dzień')
plt.ylabel('Sprzedaż')
plt.title('Sprzedaż per produkt w czasie')
plt.legend()
plt.show()  
```
