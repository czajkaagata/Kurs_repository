 Zadanie 9 – Kategorie BMI dla listy pacjentów
Masz dane pacjentów: 
wagi = [70, 85, 62, 95, 78] (kg) i 
wzrosty = [1.75, 1.80,
1.65, 1.78, 1.72] (m). Oblicz BMI dla każdego i sklasyfikuj (niedowaga < 18.5, norma
18.5-24.9, nadwaga >= 25).
Dataset:
Listy równoległe: wagi i wzrosty
Wymagania:
Użyj pętli 
for i in range(len(wagi)):
Oblicz 
bmi = wagi[i] / (wzrosty[i] ** 2)
Sklasyfikuj używając 
if/elif/else
Wyświetl wyniki dla każdego pacjenta
Oczekiwany rezultat:
5 linii z BMI i kategorią dla każdego pacjenta

```python
wagi = [70, 85, 62, 95, 78]
wzrosty = [1.75, 1.80, 1.65, 1.78, 1.72]
for i in range(len(wagi)):
    bmi = wagi[i] / (wzrosty[i] ** 2)
    if bmi < 18.5:
        kategoria = "niedowaga"
    elif 18.5 <= bmi < 25:
        kategoria = "prawidłowa masa ciała"
    elif 25 <= bmi < 30:
        kategoria = "nadwaga"
    else:
        kategoria = "otyłość"
    print(f"BMI = {bmi:.2f}, {kategoria}")
```

Zadanie 10 – Znajdź wszystkie liczby pierwsze do 100
Napisz program, który znajdzie wszystkie liczby pierwsze od 2 do 100 używając
zagnieżdżonych pętli.
Wymagania:
Pętla zewnętrzna: 
for n in range(2, 101):
Pętla wewnętrzna: sprawdź dzielniki od 2 do √n
Zmienna 
czy_pierwsza = True
Jeśli znajdziesz dzielnik, ustaw 
Jeśli 
czy_pierwsza = False i użyj 
break
czy_pierwsza == True , dodaj do listy
Oczekiwany rezultat:
Lista liczb pierwszych: [2, 3, 5, 7, 11, 13, ..., 97]
Liczba znalezionych: 25

```python
pierwsze=[]

for n in range(2, 101):
    pierwsza=True
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            pierwsza=False
            break
    if pierwsza==True:
        pierwsze.append(n)

print(f"Lista liczb pierwszych: {pierwsze}")
print(f"Liczba znalezionych: {len(pierwsze)}")
```

Zadanie 11 – Analiza trendów wzrostowych/spadkowych
Masz dane sprzedaży miesięcznej: 
[1200, 1350, 1290, 1480, 1520, 1510, 1680,
1720, 1700, 1890, 1950, 2100] . Zidentyfikuj miesiące, w których sprzedaż wzrosła w
porównaniu do poprzedniego miesiąca.
Dataset:
Lista 12 wartości (sprzedaż miesięczna w PLN)
Wymagania:
Pętla 
for i in range(1, len(sprzedaz)):
Porównaj 
sprzedaz[i] z 
sprzedaz[i-1]
Oblicz zmianę procentową
Wyświetl miesiące ze wzrostem
Oczekiwany rezultat:
Lista miesięcy ze wzrostem i procentowy przyrost

```python
dane= [1200, 1350, 1290, 1480, 1520, 1510, 1680,1720, 1700, 1890, 1950, 2100]
miesiace = ["styczeń", "luty", "marzec", "kwiecień", "maj", "czerwiec",
    "lipiec", "sierpień", "wrzesień", "październik", "listopad", "grudzień"]
lista=[]
for i in range(1, len(dane)):
    zmiana = ((dane[i] - dane[i-1]) / dane[i-1]) * 100
    if zmiana > 0:
        print(f"{miesiace[i]}, wzrost o {zmiana:.2f}%")
        lista.append((miesiace[i], f"{zmiana:.2f}%"))
print(lista)
#Oblicz zmianę procentową
#Wyświetl miesiące ze wzrostem
#Oczekiwany rezultat:
#Lista miesięcy ze wzrostem i procentowy przyrost
```
