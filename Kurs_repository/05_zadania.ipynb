```python
""" Zadanie 1 Tworzenie i indeksowanie listy
Stwórz listę zawierającą pomiary temperatury z 7 dni: 
[22.5, 23.0, 21.5, 24.0, 23.5, 22.0, 23.0] . Wyświetl:
Temperaturę z pierwszego dnia
Temperaturę z ostatniego dnia
Temperatury z trzech środkowych dni (slicing) """

data = [22.5, 23.0, 21.5, 24.0, 23.5, 22.0, 23.0]
print("Temperatura z pierwszego dnia:", data[0])
print("Temperatura z ostatniego dnia:", data[-1])
print("Temperatury z trzech środkowych dni:", data[2:5])
```

```python
""" Zadanie 2 – Dodawanie elementów do listy
Zacznij od pustej listy 
results = [] . Dodaj do niej 5 wartości accuracy: 0.75, 0.82, 0.79,
0.85, 0.88. Następnie:
Wyświetl długość listy
Wyświetl średnią wartość
Wyświetl maksymalną wartość"""
results = []
results.append(0.75)
results.extend([0.82, 0.79, 0.85, 0.88])
print("Długość listy:", len(results))
print("Średnia wartość:", sum(results) / len(results))  
print("Maksymalna wartość:", max(results))
```

```python
"""Zadanie 9 – Filtrowanie outlierów
Masz listę pomiarów: 
[23, 24, 22, 150, 23, 25, -10, 24, 23] . Outlierem jest wartość
poza zakresem [20, 30]. Stwórz nową listę bez outlierów używając list comprehension."""
pomiar = [23, 24, 22, 150, 23, 25, -10, 24, 23]
czyste_pomiar = [x for x in pomiar if 20 <= x <= 30]
list comprehension --- IGNORE ---
print("Pomiary bez outlierów:", czyste_pomiar)
```

```python
"""Zadanie 10 Masz listę transakcji (kategoria, kwota):
transactions = [
('food', 50),
('transport', 20),
('food', 30),
('entertainment', 40),
('transport', 15),
('food', 25)
]
Oblicz łączną kwotę dla każdej kategorii. Wynik zapisz w słowniku (użyj pętli i warunków)"""
transactions = [
('food', 50),
('transport', 20),
('food', 30),
('entertainment', 40),
('transport', 15),
('food', 25)
]
suma_kategorii = {}
for kategoria, kwota in transactions:
    if kategoria in suma_kategorii:
        suma_kategorii[kategoria] += kwota
    else:
        suma_kategorii[kategoria] = kwota

print("Suma kwot dla każdej kategorii:", suma_kategorii)    
```

```python
"""Zadanie 11 – Normalizacja tekstów
Masz listę kategorii produktów z różnymi formatami:
categories = ["  Electronics  ", "ELECTRONICS", "electronics", "  Books", 
"books  ", "BOOKS"]
Znormalizuj je (strip + lower) i znajdź unikalne kategorie."""
categories = ["  Electronics  ", "ELECTRONICS", "electronics", "  Books", "books  ", "BOOKS"]
cat_unikalne={}
for a in categories:
    a=a.strip().lower().replace(" ","_")
    if a in cat_unikalne:
        cat_unikalne[a]+=1
    else:
        cat_unikalne[a]=1
print("Unikalne kategorie:", list(cat_unikalne.keys()))

#LUB BEZ LICZNIKA
unikalne_set = set() # Tworzymy pusty zbiór

for a in categories:
    czysta_kat = a.strip().lower().replace(" ","_")
    unikalne_set.add(czysta_kat)

print(unikalne_set)

#from collections import Counter
# Czyścimy wszystko naraz i wrzucamy do licznika
#cat_unikalne = Counter([a.strip().lower().replace(" ","_") for a in categories])
#print(dict(cat_unikalne))
```
